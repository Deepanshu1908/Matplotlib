```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_values = [10,20,30,40,50]
y_labels = ['A','B','C','D','E']
```

```python
a = plt.pie(x_values, labels = y_labels, autopct = '%0.2f%%', radius = 2, startangle = 90, shadow = True, explode = [0,0,0,0,0.2])
plt.axis = "equal"
plt.show()
```
