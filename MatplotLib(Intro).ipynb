## Intro of MatplotLib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5]
y = [23,32,45,50,39]
```

```python
plt.plot(x,y)
```

**Properties of Plot**

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.plot(x,y,label = 'temperature',color = "Green",linewidth = 5, marker = "D", markersize = 10)
plt.legend(loc = "best", shadow = True)
plt.grid()
plt.title("Temp VS Day")
```
